Please apply train test split (%75-%25) Then grid search (hyperparameter tuning) Then classification report and confusion matrix (on the test dataset) Then compare the findings from the three algorithms

```python
data = pd.read_excel("Temporary_data3_Left_Right_Copy.xlsx")
data.head()
```

```python
Y = data["CDRGLOB"]
data.drop(columns = ["CDRGLOB"], axis = 1, inplace = True)
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(data, Y,train_size=0.75,stratify=Y)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc= DecisionTreeClassifier() 
dtc_params = { 'criterion':['gini', 'entropy'],
               'min_samples_split':[2,3,4],
               'min_samples_leaf':[1,2,3],
               'max_features':['sqrt', 'log2']}
dtc_grid = GridSearchCV(dtc, dtc_params, cv=5)
dtc_grid.fit(Xtrain, Ytrain)
dtc_grid.best_params_
grid_pred = dtc_grid.predict(Xtest)

print("Classification Report for Decision Tree Classifier:\n",classification_report(Ytest, grid_pred))
print("Confusion Matrix for Decision Tree Classifier:\n",confusion_matrix(Ytest, grid_pred))
```

```python
from sklearn.svm import SVC

svc=SVC()
svc_params= {'max_iter': [10,100,1000],
             'class_weight' : ['balanced'],
             'C': [0.1, 1, 100],
             'gamma': [1, 0.1, 0.01, 0.0001],
             }
svc_grid = GridSearchCV(svc, svc_params, refit=True)
svc_grid.fit(Xtrain, Ytrain)
svc_grid.best_params_
grid_pred = svc_grid.predict(Xtest) 

print("Classification Report for Support Vector Classification:\n",classification_report(Ytest, grid_pred))
print("Confusion Matrix for Support Vector Classification:\n",confusion_matrix(Ytest, grid_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
rfc_params = {  'n_estimators': [10, 25, 50, 100],
                'criterion': ['gini', 'entropy'],
                'max_depth': [1, 2, 3, 4, 5,7,10],
                'min_samples_split': [2, 3, 4],
              }
rfc_grid = GridSearchCV(rfc, rfc_params, refit=True)
rfc_grid.fit(Xtrain, Ytrain)
rfc_grid.best_params_
grid_pred = rfc_grid.predict(Xtest) 

print("Classification Report for Random Forest Classifier:\n",classification_report(Ytest, grid_pred))
print("Confusion Matrix for Random Forest Classifier:\n",confusion_matrix(Ytest, grid_pred))
```
