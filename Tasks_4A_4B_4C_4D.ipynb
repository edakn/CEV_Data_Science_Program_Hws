TASK-4B: Please provide the visual illustrations of the dataset that will demonstrate the insightful distributions with respect to the target column (CDRGLOB: the level of the Alzheimer Disease -AD) 

```python
x = data ["INDEPEND"]
y = data["HIPPOVOL"]
labels = data["CDRGLOB"]
plt.scatter(x, y, s=10, cmap = "viridis", c=labels, alpha=0.5)
plt.xlabel("Independence")
plt.ylabel("Hippocampus Volume")
plt.title("Independence-HippoVol Scatter")
plt.colorbar()
plt.show()
```

```python
x = data ["GRAYVOL"]
y = data["WHITEVOL"]
labels = data["CDRGLOB"]
plt.scatter(x, y, s=10, cmap = "viridis", c=labels, alpha=0.5)
plt.xlabel("Gray Volume")
plt.ylabel("White Volume")
plt.title("Gray-White Volume Scatter")
plt.colorbar()
plt.show()
```

```python
#Another Solution
data1=data['WHITEVOL']
data2=data['GRAYVOL']
corr, _ = pearsonr(data1, data2)
print('Pearsons correlation analysis between Gray Volume and White Volume variables: %.3f' % corr)
```
